```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_1 = pd.read_csv('./데이터/경기전반전망_SBHI.csv', encoding = 'cp949')
```

```python
df_1
```

```python
# 불필요한 row 삭제
for i in range(61):
    df_1.drop(i, inplace = True)
```

```python
#불필요한 columns 삭제
df_1.drop(['Unnamed: 100', '단위', '2015.01 월'], axis = 1, inplace = True)
```

```python
# sbhi 수치만 추출
df_1 = df_1[df_1['항목'] == 'SBHI']
```

```python
df_1 = df_1.reset_index(drop = True)
```

```python
df_1 = df_1.fillna(0)
```

```python
df_1 = df_1.T
```

```python
df_1.rename(columns=df_1.iloc[0],inplace=True)
```

```python
df_1 = df_1.drop(['구분별','항목'])
```

```python
df_1['시기'] = df_1.index
```

```python
df_1 = df_1.reset_index(drop = True)
```

```python
df_1.drop(['일반기업', '혁신형기업', '소기업', '중기업'], axis = 1 ,inplace = True)
```

```python
df_1.to_csv('경기전반_sbhi.csv', encoding = 'cp949')
```
